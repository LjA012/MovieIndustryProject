## Findings
> We found that runtime and overall budget did not have as large of an impact on a movie's success as we expected.  For example, on average, we found that to have a noticable impact on a film's rating, one would have to increase the budget by \\$200 million dollars. Thus first and foremost, the movie should be made right without worrying about cost and runtime.  That being said, our recommendation is to stay within 1 standard deviation of the mean for both budget and runtime.  

> On the other hand, we found that the Genre, Actor/Actress and Director had a bigger influence on a movies success.  The top Genres for a successful movie were Action, Adventure, Musical and SciFi.  Out of the 20 most common Actors and Directors, we chose the ones that resulted in a higher earnings average, then on the ratings average - which is how Microsoft defines success.

## Recommendations

* Genres
    * SciFi
    * Adventure
* Directors
    * Christopher Nolan
    * Francis Lawrence
* Actors
    * Jennifer Lawrence
    * Robert Downey Jr.


> Our findings conclude that if Microsoft wants to create a successful movie, we recommend choosing the Genre to be SciFi, Adventure or both; with either Christopher Nolan or Francis Lawrence directing the film and the cast consisting of Jennifer Lawrence and Robert Downey Jr.  As far production and budget are concerned, recommend the runtime to come in between 90 and 125 minutes, and we recommend a production budget no higher than \\$110 million dollars.  These recommendations will yield a successful moive with high ratings and high earnings.
